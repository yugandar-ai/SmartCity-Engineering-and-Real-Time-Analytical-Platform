```python
import pandas as pd
import mysql.connector


# 1. LOAD CSV (SAFE MODE)


file_path = r"D:\major project 4-2\2.csv files codes\Hyderabad public saftey\public saftey python code\hyderabad_public_safety_detailed.csv"

df = pd.read_csv(
    file_path,
    on_bad_lines="skip",   # skip wrong rows
    engine="python"
)

df = df.where(pd.notnull(df), None)

print("Dataset Loaded")
print(df.head())
print("Total rows:", len(df))



# 2. CONNECT MYSQL


conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2711",
    database="smart_city"
)

cursor = conn.cursor()

print("Connected to MySQL")


# 3. INSERT QUERY


insert_query = """
INSERT INTO hyderabad_public_safety (

incident_id,
date,
time,
year,
month,
hour,
day_of_week,

area,
locality,
street_name,
zone,
location_type,

latitude,
longitude,

incident_type,
incident_sub_type,

severity,
status,
department,

response_time_minutes,

people_involved,
victim_count,
injuries,

damage_cost,

report_source,
lighting_condition,
cctv_available,
premises_status,
weather_condition,
offender_captured

)

VALUES (%s,%s,%s,%s,%s,%s,%s,
        %s,%s,%s,%s,%s,
        %s,%s,
        %s,%s,
        %s,%s,%s,
        %s,
        %s,%s,%s,
        %s,
        %s,%s,%s,%s,%s,%s)
"""



# 4. INSERT LOOP (SAFE)


for _, row in df.iterrows():

    values = (

        row.get("incident_id"),
        row.get("date"),
        row.get("time"),
        row.get("year"),
        row.get("month"),
        row.get("hour"),
        row.get("day_of_week"),

        row.get("area"),
        row.get("locality"),
        row.get("street_name"),
        row.get("zone"),
        row.get("location_type"),

        row.get("latitude"),
        row.get("longitude"),

        row.get("incident_type"),
        row.get("incident_sub_type"),

        row.get("severity"),
        row.get("status"),
        row.get("department"),

        row.get("response_time_minutes"),

        row.get("people_involved"),
        row.get("victim_count"),
        row.get("injuries"),

        row.get("damage_cost"),

        row.get("report_source"),
        row.get("lighting_condition"),
        row.get("cctv_available"),
        row.get("premises_status"),
        row.get("weather_condition"),
        row.get("offender_captured")

    )

    try:
        cursor.execute(insert_query, values)
    except:
        continue


conn.commit()

print("Data inserted successfully")


cursor.close()
conn.close()
```
