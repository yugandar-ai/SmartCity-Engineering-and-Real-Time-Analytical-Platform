```python
import pandas as pd
import mysql.connector


# =========================
# 1. load csv file
# =========================

file_path = r"D:\major project 4-2\indian data sets without using api\traffic bengalure\benguluru data set\bengaluru_traffic_monitoring_dataset_final.csv"

df = pd.read_csv(file_path)

df = df.where(pd.notnull(df), None)

print("Dataset Loaded")
print(df.head())
print("Total rows:", len(df))


# =========================
# 2. connect to mysql (fixed)
# =========================

# connect to mysql database
conn = mysql.connector.connect(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="2711",
    database="smart_city",
    autocommit=False
)

cursor = conn.cursor()

# check database
cursor.execute("SELECT DATABASE();")
print("Connected DB:", cursor.fetchone())


# =========================
# 3. insert query
# =========================

insert_query = """
INSERT INTO traffic_bengaluru_csv (

date,
time,
area_name,
zone,
road_name,

latitude,
longitude,

traffic_volume,
average_speed,
travel_time_index,

congestion_level,
peak_hour,
holiday,
event_impact,

road_type,
lane_count,
road_capacity,
road_capacity_utilization,

signal_waiting_time,

incident_reports,
accident_reports,
emergency_vehicle_count,
emergency_response_time,

roadwork_activity,
road_condition_score,

bus_count,
auto_count,
car_count,
bike_count,
truck_count,
ev_count,

public_transport_usage,
metro_passenger_count,
distance_to_metro,

parking_usage,
parking_capacity,

pedestrian_count,
cyclist_count,

weather,
temperature,
humidity,

air_quality_index,
environmental_impact,
fuel_consumption,
co2_emission,
noise_level,

signal_compliance,

cctv_count,
police_count,

data_source

)

VALUES (%s,%s,%s,%s,%s,
        %s,%s,
        %s,%s,%s,
        %s,%s,%s,%s,
        %s,%s,%s,%s,
        %s,
        %s,%s,%s,%s,
        %s,%s,
        %s,%s,%s,%s,%s,%s,
        %s,%s,%s,
        %s,%s,
        %s,%s,
        %s,%s,%s,
        %s,%s,%s,%s,%s,
        %s,
        %s,%s,
        %s)
"""


# =========================
# 4. insert loop (same logic)
# =========================

for _, row in df.iterrows():
    try:
        cursor.execute(insert_query, tuple(row))
    except:
        continue


conn.commit()

print("Data inserted successfully!")


# check rows count
cursor.execute("SELECT COUNT(*) FROM traffic_bengaluru_csv")
print("Rows:", cursor.fetchone())


cursor.close()
conn.close()
```
